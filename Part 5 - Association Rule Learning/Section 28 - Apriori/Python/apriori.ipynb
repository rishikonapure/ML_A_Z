```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data Preprocessing

```
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None) #header is set to none because out data set dont have any column name.
#apriori expects the dataset in form of list
#creating list of trnsactions
transactions = [] #intialize empty list
for i in range(0, 7501): 
  transactions.append([str(dataset.values[i,j])for j in range(0, 20)])
  #to get all the transactions
#dataset has 20 columns i.e upto 20 products in each transaction
#for loop to take all products from each transaction
#products should be in string format

```

## Training the Apriori model on the dataset

```
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_lenght = 2, max_lenght = 2)

#min_support = 3*7/7501 3 - one product in min 3 transactiosns per day , 7 -because 7501 transactions recorded over a week, 7501 - total number of transactions
#min_confidence - this value can be set on the basis of business requirment.
#lift - choose minimum lift of 3
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```
results = list(rules)
results
```

### Putting the results well organised into a Pandas DataFrame

```
def inspect(results):
  lhs         = [tuple(result[2][0][0])[0] for result in results]
  rhs         = [tuple(result[2][0][1])[0] for result in results]
  supports    = [result[1] for result in results]
  confidences = [result[2][0][2] for result in results]
  lifts       = [result[2][0][3] for result in results]
  return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
